# Creating and Annotating Line Charts 

```python
import matplotlib.pyplot as plt # to import the libaries to create charts 
savings = [0, 50, 100, 150,200, 250, 300, 250, 200, 300, 250, 500] # creating the y series for the chart
months = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"] # creating a list for x series
plt.plot (months, savings) # specify the x and y series, if x is not specified, it will be autmatically created 
plt.xlabel('Month') # to create the label for the x series
plt.ylabel('Savings in $') # to create the label for the y series
plt.title('Saving Account Results ') # to create a title for the chart
plt.show() # to print the chart and should always be the final line of code
```

## Line Charts -  MATLAB Method 

```python
import matplotlib.pyplot as plt # to import the libaries to create charts 
#%matplotlib inline # specifies to print chart within the cell
# Set the x-axis to a list of strings for each month and assign it a label/series name
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
plt.xlabel('Month')
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month and give it a label
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.ylabel('Fare in USD')
# Create the plot
plt.plot(x_axis, y_axis)
plt.show() # can also print the chart and should always be the final line of code
```

```python
# Create the plot specifying the axis values, markers, colour, line width and label for the line 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

## Line Charts -  Object Orientated Method 

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
# plt.subplots() returns a tuple (list with unchangeable variables which are fig and ax)
# fig references the figure and ax references the axes in the figure
ax.plot(x_axis, y_axis) # plots the x and y axis we've defined above 
```

```python
# Create the plot with ax.plt()
fig = plt.figure() # creates the figure/chart area
ax = fig.add_subplot() # creates the axes for the chart
ax.plot(x_axis, y_axis) # populates the chart by defining what is in the figure
# Change the line color to green, Change the marker to a diamond, Change the line width to 2 points, Add a legend for Boston.
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label="Boston")
# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add a y-axis limit.
ax.set_ylim(0, 45)
# Add grid lines
ax.grid()
```

# Creating and Annotating Bar Charts

## Bar Charts - Matlab Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot, you can change the axis around to change which one prints where 
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# gca() gets the current axes and the .invert inverts/flips the sorting order of the axes (otherwise it prints Dec first)
plt.gca().invert_yaxis()
# Create the title, axes labels, and legend
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.legend()
```

## Bar Charts - Object Orientated Method

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
# Change the bar color to cyan, Add a legend for the city of Chicago.
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
# Add a title, axes labels and legend 
ax.set_title('Pyber Fare by Month')
ax.set_xlabel('Fares($)')
ax.set_ylabel('Months')
ax.legend()
# Invert the y-axis data.
plt.gca().invert_yaxis()
```

# Scatter Plots 

## Scatter Plots - Matlab Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Change the color of the markers to red.
# Add a legend for the city of Chicago.
# Switch the axis so that the Fare($) data is on the x-axis.
plt.plot(y_axis, x_axis, 'o', c='red',label = 'Chicago')

# Add a title and axes labels.
plt.title("Fares by Month")
plt.xlabel("Fares ($)")
plt.ylabel("Months")
plt.legend()
         
# Add a limit for the x-axis data.
plt.xlim (0,45)

# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# In order to create a bubble chart, we will create a list 
# We will loop through the y axis values and append a value of y axis * 3 to the list we created
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
# We can create bubble charts by making the size of the bubble = to the value using s = the list we created 
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Alternatively, we can use a list comprehension which replaces loops 
# The syntax is new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Scatter Plot - Object Orientated Method

```python
fig, ax = plt.subplots()
# Switch the axis so that the Fare($) data is on the x-axis.
# Change the color of the markers to sky blue, done by defining the c or colour value
# Change the size of the markers to 5 times each data point, done by using a list comprehenstion
# Make the color 20% transparent, done by assigning an alpha value between 0 to 1
# Add a black edge color to the circles, done by defining the edgecolour value
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
ax.scatter(y_axis, x_axis, c='skyblue', s = [i * 5 for i in y_axis], alpha = 0.2, edgecolors = 'black', linewidth = 2, label = 'Boston')
# Add a title, labels, legend, limit, and invert the  axis
ax.set_title('Fares per Month')
ax.set_ylabel('Months')
ax.set_xlabel('Fares($)')
ax.legend()
ax.set_xlim(0,45)
plt.gca().invert_yaxis()
```

# Pie Charts

## Pie Charts - MATLAB Method

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Exploding values allows us to specify from 0 to 1 how far from the center to explode from 
explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
# the values in the lists above correspond to the index/list of our axes 

# This creates the figure/subplot size for our pie to be printed into 
plt.subplots(figsize=(8, 8))
# This creates the pie chart and defines what to explode, assign colours, add labels, and formats the % printed to 1 decimal
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

## Pie Charts - Object Orientated Method

```python
#Increase the figure size to 8x8.
fig, ax = plt.subplots(figsize = (8,8))
# List for explode and colours 
# Explode the two highest percentage months.
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis,labels=x_axis, 
       autopct='%.1f%%', # Add a percentage to one decimal place to each wedge of the pie.
       explode=explode_values, # Explode the two highest percentage months.
       colors=colors, # Add new colors of your own choosing or use the colors from the previous pie chart.
       shadow=True, # Add a shadow.
       startangle=90, # Add a start angle so that January is at the top.
       counterclock=False) # Reverse the order so that the month order is in a clockwise direction.
       
plt.show()
```
