```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Using the .plot function we can just create a line graph
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# We can create a bar graph from the same data using .bar or kind
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='pie')
```

```python
# To be able to set x-axis and tick locations.

# With the NumPy module, we create an array of numbers using the arange function for the length of the pyber_ride_df
x_axis = np.arange(len(pyber_ride_df))
# Then we set the tick locations for every value, or month, in the x-axis using list comprehension
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
# We add the xticks() function and pass in the tick_locations and the labels of the x-ticks, pyber_ride_df["Month"]
plt.xticks(tick_locations, pyber_ride_df["Month"])
# Print the chart 
plt.show()
```
